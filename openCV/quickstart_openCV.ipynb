```python
import cv2
```

```python
img = cv2.imread("pikachu.jpg")
gray_img = cv2.imread("pikachu.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
#to correct the color channels of image b/w matplotlib and opencv
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
plt.axis('off')
```

```python
#color space change is not required when reading and dislaying in opencv itself
cv2.imshow('image',img)
cv2.imshow('gray scale', gray_img)
cv2.waitKey(100) #waits 100ms before closing the window
# cv2.waitKey(0) #stops displaying when any key is pressed
# cv2.destroyAllWindows() 
```

```python
print(img.shape)
```

```python
cropped_img = img[50:150, 100:200, :] #cropping/ slicing
plt.imshow(cropped_img)
plt.show()
```

```python
cropped_img.shape
```

```python
total_pixels = img.shape[0]*img.shape[1]
total_pixels
```

```python
#Resizing images
crop_img_2 = cv2.resize(cropped_img,(150,150))
plt.imshow(crop_img_2)
plt.title("Bulbasaur")
plt.show()
```

```python
crop_img_3 = cv2.resize(cropped_img,(50,50))
plt.imshow(crop_img_3)
plt.title("Bulbasaur part 2")
plt.show()
```
